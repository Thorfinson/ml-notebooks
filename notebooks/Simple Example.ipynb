# Altair example on Binder

imports need to be in the requirements.txt. Jupyter lab is auto installed

```python
import pandas as pd
import numpy as np
import altair as alt
from sklearn.datasets import make_blobs
```

Create three cluster with two features

```python
X, y = make_blobs(n_samples=200, centers=3, n_features=2, random_state=0)
```

Make a Dataframe for altair

```python
df = pd.concat([pd.DataFrame(X), pd.DataFrame(y)], axis=1)
df.columns =['x','y','cluster']
```

visualize as a scatter plot

```python
alt.Chart(df).mark_circle().encode(x='x', y='y', color='cluster:N')
```
